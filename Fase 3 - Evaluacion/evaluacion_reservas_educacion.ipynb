```python
#importamos las galerías necesarias
import numpy as np
import pandas as pd

#configuramos para ver todas las columnas
pd.set_option("display.max_columns", None)
```

```python
#cargamos el dataset
df_evaluacion = pd.read_csv("/Users/lucianayelicich/Desktop/Adalab/bda-modulo-3-evaluacion-final-lucianayelicich/Data/archivo_aviacion_final.csv", index_col= 0)
df_evaluacion.head()
```

```python
# Media de todas las columnas numéricas agrupadas por educación
df_evaluacion.groupby("Education").mean(numeric_only=True).T # transpose
```

```python
#filtramos el conjunto de datos para incluir únicamente las columnas relevantes: 'Flights Booked' y 'Education', y calculamos las estadísticas descriptivas básicas
educacion_reservas = df_evaluacion.groupby("Education")["Flights Booked"].agg(["mean", "median", "std", "var", "min", "max"])
educacion_reservas 
```

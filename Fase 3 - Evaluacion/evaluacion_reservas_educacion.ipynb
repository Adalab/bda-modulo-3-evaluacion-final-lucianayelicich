```python
#importamos las galer√≠as necesarias
import numpy as np
import pandas as pd

#configuramos para ver todas las columnas
pd.set_option("display.max_columns", None)
```

```python
#cargamos el dataset
df_evaluacion = pd.read_csv("/Users/lucianayelicich/Desktop/Adalab/bda-modulo-3-evaluacion-final-lucianayelicich/Data/archivo_aviacion_final.csv")
df_evaluacion.head()
```

```python
#obtenemos la media de todas las columnas num√©ricas agrupadas por educaci√≥n
df_evaluacion.groupby("Education").mean(numeric_only=True).T #hacemos transpose para verlas con claridad
```

```python
#filtramos el conjunto de datos para incluir √∫nicamente las columnas relevantes: 'Flights Booked' y 'Education', y calculamos las estad√≠sticas descriptivas b√°sicas
educacion_reservas = df_evaluacion.groupby("Education")["Flights Booked"].agg(["mean", "median", "std", "var", "min", "max"])
educacion_reservas 
```

```python
#guardamos una copia del dataframe en una nueva variable para efectuar la limpieza
df_final = df_aviacion.copy()
df_final.head()
```

```python
#revisamos valores nulos absolutos
df_final.isnull().sum()
```

```python
#dividimos la suma de valores nulos entre el n√∫mero de filas del df y multiplicamos por 100 para sacar el porcentaje o frequencia relativa de nulos y los ordenamos
round((df_final.isnull().sum() / df_final.shape[0]) * 100,2).sort_values(ascending=False)
```

```python
#debido a que el mes y el a√±o de cancelaci√≥n tienen muchos datos NaN, pero que en el resumen brindado se aclara que los datos se
#brindan solo en caso de corresponder, interpretamos que dichas membres√≠as est√°n vigentes y no han sido canceladas
#por lo que los mismos no ser√°n imputados
```

```python
#obtenemos las principales estad√≠sticas del df, para chequear datos at√≠picos o nulos
df_final.describe().T
```

```python
#observamos que la columna Salary tiene valores negativos, lo cual es imposible y se trata de un dato incoherente
#chequeamos esa columna en particular para cerciorarnos que se trata de un error en la recolecci√≥n del dato
df_final["Salary"].agg(["mean", "median", "std", "var", "min", "max"])
```

```python
#debemos quitar el gui√≥n de la columna Salary, para lo cual primero debemos transformarla a tipo str, para aplicar el replace
df_final['Salary'] = df_final['Salary'].astype(str)

#reemplazamos el guion "-" por una cadena vac√≠a, lo que remueve el guion de cualquier valor en la columna.
df_final['Salary'] = df_final['Salary'].str.replace('-', '', regex=False)

#convertimos la columna de vuelta a tipo num√©rico, y le indicamos que lo que no pueda corregir lo transforme a NaN
df_final['Salary'] = pd.to_numeric(df_final['Salary'], errors='coerce')

print("La columna 'Salary' ha sido limpiada y ahora tiene un formato num√©rico.")
print(df_final['Salary'].head())
```

```python
#decidimos imputar los valores nulos de la columna Salary usando IterativeImputer porque es una estrategia de imputaci√≥n avanzada 
#que preserva las relaciones y dependencias entre las variables. A diferencia de m√©todos simples como la media o la mediana, 
#IterativeImputer considera el valor de otras columnas para hacer una estimaci√≥n m√°s precisa
print("\n--- Imputaci√≥n con m√©todos avanzados (IterativeImputer) ---")
# Haremos una copia del DataFrame para aplicar ambos m√©todos y comparar.
df_final_copia = df_final.copy()

#definimos las columnas a usar para la imputaci√≥n, que son columnas num√©ricas relacionadas con el salario
columnas_para_imputar = ["Salary", "CLV", "Flights Booked", "Points Accumulated"] 

#ajustamos el imputador con todas las columnas seleccionadas
imputer_iterative = IterativeImputer(max_iter=20, random_state=42)
imputado_iterative = imputer_iterative.fit_transform(df_final_copia[columnas_para_imputar])

# La primera columna del resultado imputado es la de "Salary"
df_final_copia["Salary_iterative"] = imputado_iterative[:, 0]
print(f"Despu√©s del 'IterativeImputer' tenemos: \n{df_final_copia[['Salary_iterative']].isnull().sum()} nulos")
```

```python
#chequemos c√≥mo queda el df
df_final_copia.head()
```

```python
#comparamos los resultados de IterativeImputer y la columna original
print("\n--- Comparando estad√≠sticos de la columna original y la imputada ---")
# Compararemos los estad√≠sticos (media, mediana, etc.) antes y despu√©s de la imputaci√≥n.
# Nos quedaremos con el m√©todo que menos modifique los estad√≠sticos.
df_final_copia.describe()[["Salary", "Salary_iterative"]]
```

```python
#eliminamos la columna 'Salary' original (la que tiene NaN)
df_final_copia = df_final_copia.drop(columns=['Salary'])

#renombramos la columna 'Salary_iterative' a 'Salary'
df_final_copia.rename(columns={"Salary_iterative": "Salary"}, inplace=True)

#reordenamos las columnas
columnas = df_final_copia.columns.tolist()
columnas.remove('Salary')

#creamos un indice con los titulos de las columnas
posicion_educacion = columnas.index('Education')

#le indicamos al df que debe insertar la columna salary nuevamente en su lugar original, despues de education
columnas.insert(posicion_educacion + 1, 'Salary')

#asignamos el nuevo orden al DataFrame
df_final_copia = df_final_copia[columnas]

print("Columnas despu√©s de la reorganizaci√≥n:")
print(df_final_copia.columns.tolist())
```

```python
#chequeamos el df
df_final_copia.head()
```

```python
#del an√°lisis EDA que realizamos, nos percatamos que el mes y a√±o son valores de tipo num√©rico y no de fecha, por lo que debemos corregirlo
#para ello primero hay que unir ambbos datos en una sola columna
#creamos la columna 'Date' y la convertimos a tipo datetime
df_final_copia['Date'] = pd.to_datetime(df_final_copia['Month'].astype(str) + '-' + df_final_copia['Year'].astype(str))

#indicamos que la nueva columna se inserte en el lugar original del df
df_final_copia.insert(loc=1, column='Date', value=df_final_copia.pop('Date'))

#verificamos que la conversi√≥n haya funcionado
print("Tipo de dato de la columna 'Date':", df_final_copia['Date'].dtype)

df_final_copia.head()
```

```python
#eliminamos las columnas year y month por ser redundantes e innecesarias
df_final_copia.drop(columns=["Year", "Month"], axis=1, inplace=True)
```

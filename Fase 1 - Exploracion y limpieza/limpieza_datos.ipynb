```python
# 8. Imputación de "Salary" con IterativeImputer y KNNImputer
print("\n--- Imputación con métodos avanzados (IterativeImputer y KNNImputer) ---")
# Haremos una copia del DataFrame para aplicar ambos métodos y comparar.
df_final_copia = df_final.copy()

# a. IterativeImputer
# Utiliza un modelo de regresión para estimar valores faltantes basándose en otras columnas.
# max_iter: número de iteraciones. random_state: para reproducibilidad.
imputer_iterative = IterativeImputer(max_iter = 20, random_state = 42)
# Ajustar y transformar los datos para las columnas especificadas.
imputer_iterative_imputado = imputer_iterative.fit_transform(df_final_copia[["Salary"]])

# Añadir las columnas imputadas al DataFrame copia con nuevos nombres
df_final_copia["Salary_iterative"] = imputer_iterative_imputado
print(f"Después del 'IterativeImputer' tenemos: \n{df_final_copia[['Salary_iterative']].isnull().sum()} nulos") 

# b. KNNImputer
# Basado en el algoritmo de k-vecinos más cercanos (KNN). Estima valores faltantes usando observaciones similares.
# n_neighbors: número de vecinos a considerar.
print("\nPreparando KNNImputer (puede tardar un poco)...")
imputer_knn = KNNImputer(n_neighbors = 5) 
# Ajustar y transformar los datos para las mismas columnas.
imputer_knn_imputado = imputer_knn.fit_transform(df_final_copia[["Salary"]]) 

# Añadir las columnas imputadas al DataFrame copia con nuevos nombres
df_final_copia["Salary_knn"] = imputer_knn_imputado 
print(f"Después del 'KNNImputer' tenemos: \n{df_final_copia[['Salary_knn']].isnull().sum()} nulos") 
```

```python
# 9. Comparar los resultados de IterativeImputer y KNNImputer
print("\n--- Comparando estadísticos de las columnas originales y las imputadas ---")
# Compararemos los estadísticos (media, mediana, etc.) antes y después de la imputación.
# Nos quedaremos con el método que menos modifique los estadísticos.
df_final_copia.describe()[["Salary","Salary_iterative", "Salary_knn"]]
```

```python
df_final_copia.head()
```

```python
print("\n--- Limpiando el DataFrame final ---")
# Eliminamos las columnas originales con nulos y las imputaciones menos óptimas.
df_final_copia.drop(["Salary", # Originales con nulos
              "Salary_iterative" # Podríamos haber elegido cualquiera de los dos, eliminamos estos
                ], axis = 1, inplace = True)

# Cambiar el nombre de las columnas restantes a sus nombres originales
nuevo_nombre = {"Salary_knn": "Salary"} 
df_final_copia.rename(columns = nuevo_nombre, inplace = True) 

```

```python
df_final_copia.head()
```

```python
# Crea la columna 'date'
# Une las columnas 'Month' y 'Year' y las convierte a un formato de fecha.
# Esto crea una columna llamada 'date' en la última posición del DataFrame.
df_final_copia['Date'] = pd.to_datetime(df_final_copia['Month'].astype(str) + '-' + df_final_copia['Year'].astype(str), format='%m-%Y')

# Formatea la columna 'date' para que solo muestre el mes y el año.
# Esto convierte el tipo de dato de la columna a string.
df_final_copia['Date'] = df_final_copia['Date'].dt.strftime('%m-%Y')

# Ahora, mueve la columna 'date' a la posición deseada si no la quieres al final.
# Usamos .insert() para colocar la columna 'date' en la posición 1.
# El argumento 'value' debe ser la columna de datos real que acabas de crear.
df_final_copia.insert(loc=1, column='Date', value=df_final_copia.pop('Date'))

df_final_copia.head()
```

```python
df_fidelidad = pd.read_csv("/Users/lucianayelicich/Desktop/Adalab/bda-modulo-3-evaluacion-final-lucianayelicich/Data/Customer Loyalty History.csv", index_col= 0)
df_fidelidad.head()
```

```python
#unimos los dataframes con merge

df_aviacion = df_vuelos.merge(right=df_fidelidad, how="left", left_on="Loyalty Number", right_on="Loyalty Number")
df_aviacion.head()
```

```python
#unimos los dataframe con join
df_aviacion2 = df_vuelos.join(df_fidelidad, on="Loyalty Number")
df_aviacion2.head()
```

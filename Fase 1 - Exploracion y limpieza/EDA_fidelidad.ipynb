```python
#exploramos los valores únicos y la frequencia absoluta de éstos para cada columna, incluidos los nulos
for columna in df.columns:
    print(f"--- Frecuencias para la columna: {columna} ---")
    print(df[columna].value_counts(dropna=False))
    print("\n")
```

```python
#exploramos los valores únicos y la frequencia relativa de éstos para cada columna, incluidos los nulos
for columna in df.columns:
    print(f"--- Frecuencias para la columna: {columna} ---")
    print(round(df[columna].value_counts(dropna = False, normalize = True)*100), 2)
    print("\n")
```

```python
#chequeamos los nulos absolutos que informa .info
df.isnull().sum()
```

```python
#chequeamos los nulos relativos que informa .info
round(df.isnull().sum()/df.shape[0] *100, 2)
```
